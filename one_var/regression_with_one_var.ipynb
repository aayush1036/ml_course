```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:,:-1].values #everything except the last column
y = dataset.iloc[:,-1].values #last column
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2,random_state=0)
```

```
from sklearn.linear_model import LinearRegression #import the linear regression class from sklearn.linear_model
regressor = LinearRegression() #make an instance of LinearRegression()
regressor.fit(X_train, y_train) #call the fit method from the instance to do linear regression
```

```
regressor.predict(X_test) #predicting y values from test set
```

```
fig, ax = plt.subplots(figsize=(16,6))
ax.scatter(X_train, y_train, color='red', label='Training set data points')
ax.plot(X_train, regressor.predict(X_train), color='blue', label='Our linear regression prediction')
ax.set_title('Regression on training set')
ax.set_xlabel('Years of experience')
ax.set_ylabel('Salary')
ax.legend(loc=0);
```

```
fig, ax = plt.subplots(figsize=(16,6))
ax.scatter(X_test, y_test, color='red', label='Test set data points')
ax.plot(X_test, regressor.predict(X_test), color='blue', label='Our linear regression prediction')
ax.set_title('Regression on test set')
ax.set_xlabel('Years of experience')
ax.set_ylabel('Salary')
ax.legend(loc=0);
```

```
x = np.array([20.7]).reshape(-1,1)
```

```
regressor.predict(x)
```

```
print(f'The coefficient of linear regression in {regressor.coef_} and the intercept is {regressor.intercept_}')
```

```
print(f'Hence the line becomes\ny={regressor.coef_}*x + {regressor.intercept_}')
```

```
coef = regressor.coef_
intrcept = regressor.intercept_
def predict(exp):
    return (coef*exp)+intrcept
```

```
predict(20.7)
```
